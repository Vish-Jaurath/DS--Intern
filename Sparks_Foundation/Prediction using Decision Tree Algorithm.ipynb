### Prediction using Decision Tree Algorithm

####  In this project I've built Decision Tree Classifier for predicting the right class.

```python
# Importing all libraries required for this project.

import pandas as pd
import numpy as np    
import matplotlib.pyplot as plt  
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

### Importing and Preprocessing the data

```python
# Reading data from remote link.

df = pd.read_csv('iris.csv')
df.head()
```

```python
import seaborn as sns
sns.set()
sns.pairplot(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']],
             hue="Species", diag_kind="kde")
```

```python
df.Species.unique()
```

```python
df['Species']= df['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df.head()
```

```python
df.tail()
```

```python
X= df.drop(['Id','Species'], axis=1)
y= df.Species
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state= 3)
```

```python
len(X_train), len(X_test)
```

### Creating a Classifier

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
clf= DecisionTreeClassifier()
clf.fit(X_train, y_train)
predictions= clf.predict(X_test)
```

```python
from sklearn import metrics
print('Accuracy of Decision Tree Classifier is:', metrics.accuracy_score(y_test, predictions))
```

### Confusion matrix

```python
cm= metrics.confusion_matrix(y_test, predictions)
cm
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
plt.figure(figsize=(15,10), dpi=70, facecolor= 'tomato')
plot_tree(clf, filled= True)
plt.title('Decision Tree', fontsize=20)
```
